# 1. Spark Job Failed - OutOfMemory Error (OOM)

## Error

```text
java.lang.OutOfMemoryError: Java heap space
Executor Lost
Container killed due to exceeding memory limits
```

## Possible Causes

- Large shuffle operations
- Using `collect()` on huge datasets
- Incorrect partitioning
- Insufficient executor memory
- Excessive caching

## Investigation

### Check Spark UI

- Jobs
- Stages
- Executors
- Storage
- SQL Tab

### Check Logs

- Driver Logs
- Executor Logs
- stderr
- stdout
- YARN/Event Logs

## Immediate Fix

Increase Executor Memory

```python
spark.conf.set("spark.executor.memory", "8g")
```

or

```bash
--executor-memory 8G
```

Avoid

```python
df.collect()
```

Use

```python
df.show()
```

## Permanent Solution

- Use `repartition()`
- Use `coalesce()`
- Broadcast small tables
- Avoid unnecessary caching
- Optimize transformations

# 2. Executor Lost Failure

- Memory overflow

- Spark UI

# 3. Shuffle Failure

## Error

```text
FetchFailedException
MetadataFetchFailedException
```

## Causes

- Huge joins
- groupBy()
- distinct()
- orderBy()

## Investigation

Spark UI

Check

- Shuffle Read
- Shuffle Write
- Failed Stage

## Solution

Increase shuffle partitions

```python
spark.conf.set("spark.sql.shuffle.partitions", "500")
```

Broadcast smaller dataframe

```python
from pyspark.sql.functions import broadcast

result = fact_df.join(broadcast(dim_df), "id")
```

# 4. Data Skew

## Symptoms

One executor runs for 30 minutes while others finish in seconds.

## Cause

One key contains most records.

Example

```text
India      95%
USA         3%
UK          2%
```

## Solution

### Salting

```python
from pyspark.sql.functions import rand

df = df.withColumn("salt", (rand()*10).cast("int"))
```

### Broadcast Join

```python
broadcast(dim_df)
```

### Repartition

```python
df.repartition(500)
```

# Best Practices for Production Support

- Never use `collect()` on large datasets.
- Monitor Spark UI regularly.
- Partition data appropriately.
- Use broadcast joins for small lookup tables.
- Maintain checkpoints for incremental loads.
- Implement retry mechanisms for transient failures.
- Validate schema before processing.
- Add data quality checks (nulls, duplicates, record counts).
- Archive processed and rejected files.
- Monitor cluster health and resource utilization.
- Document Root Cause Analysis (RCA) after every production incident.
- Build idempotent pipelines to safely rerun failed jobs.
- Configure alerts for job failures and SLA breaches.
