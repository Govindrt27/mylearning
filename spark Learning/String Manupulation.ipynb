String manipulation is one of the most frequently used topics in **Hive, Spark SQL, and Standard SQL**, especially in ETL pipelines, data cleansing, and interview questions.

---

# String Manipulation Functions Comparison

| Purpose              | Standard SQL                    | Hive                       | Spark SQL                  | Notes                                  |
| -------------------- | ------------------------------- | -------------------------- | -------------------------- | -------------------------------------- |
| Convert to Uppercase | `UPPER()`                       | `upper()`                  | `upper()`                  | Same                                   |
| Convert to Lowercase | `LOWER()`                       | `lower()`                  | `lower()`                  | Same                                   |
| String Length        | `CHAR_LENGTH()` / `LENGTH()`    | `length()`                 | `length()`                 | Hive/Spark use `length()`              |
| Trim Spaces          | `TRIM()`                        | `trim()`                   | `trim()`                   | Same                                   |
| Left Trim            | `LTRIM()`                       | `ltrim()`                  | `ltrim()`                  | Same                                   |
| Right Trim           | `RTRIM()`                       | `rtrim()`                  | `rtrim()`                  | Same                                   |
| Concatenate          | `CONCAT()` / `\|\|`             | `concat()`                 | `concat()`                 | Hive doesn't support `\|\|` by default |
| Substring            | `SUBSTRING()`                   | `substr()` / `substring()` | `substr()` / `substring()` | Both supported                         |
| Find Position        | `POSITION()`                    | `instr()`                  | `instr()`                  | Different syntax                       |
| Replace String       | `REPLACE()`                     | `replace()`                | `replace()`                | Same                                   |
| Regex Replace        | `REGEXP_REPLACE()`              | `regexp_replace()`         | `regexp_replace()`         | Same                                   |
| Regex Extract        | `REGEXP_SUBSTR()` *(Oracle)*    | `regexp_extract()`         | `regexp_extract()`         | Different names                        |
| Split String         | Database-specific               | `split()`                  | `split()`                  | Returns array                          |
| Join Array           | `STRING_AGG()` *(DB-specific)*  | `concat_ws()`              | `concat_ws()`              | Different purpose                      |
| Left Padding         | `LPAD()`                        | `lpad()`                   | `lpad()`                   | Same                                   |
| Right Padding        | `RPAD()`                        | `rpad()`                   | `rpad()`                   | Same                                   |
| Reverse String       | Database-specific               | `reverse()`                | `reverse()`                | Same in Hive/Spark                     |
| Repeat String        | `REPEAT()` *(DB-specific)*      | `repeat()`                 | `repeat()`                 | Same                                   |
| ASCII Value          | `ASCII()`                       | `ascii()`                  | `ascii()`                  | Same                                   |
| Character from ASCII | `CHR()`                         | `chr()`                    | `char()` / `chr()`         | Version-dependent in Spark             |
| Init Cap             | `INITCAP()` *(Oracle/Postgres)* | `initcap()`                | `initcap()`                | Capitalize words                       |
| Translate Characters | `TRANSLATE()`                   | `translate()`              | `translate()`              | Character mapping                      |
| String Format        | `FORMAT()` *(DB-specific)*      | `printf()`                 | `printf()`                 | Similar functionality                  |
| Soundex              | `SOUNDEX()` *(DB-specific)*     | ❌                          | `soundex()`                | Spark supports it                      |
| Base64 Encode        | Database-specific               | `base64()`                 | `base64()`                 | Encode string                          |
| Base64 Decode        | Database-specific               | `unbase64()`               | `unbase64()`               | Decode string                          |

# 1. UPPER()

Convert to uppercase.

### Standard SQL

```sql
SELECT UPPER('openai');
```

### Hive

```sql
SELECT upper('openai');
```

### Spark SQL

```sql
SELECT upper('openai');
```

**Output**

```
OPENAI
```

---

# 2. LOWER()

```sql
SELECT lower('OPENAI');
```

Output

```
openai
```

---

# 3. LENGTH()

```sql
SELECT length('ChatGPT');
```

Output

```
7
```

---

# 4. TRIM()

Remove leading and trailing spaces.

```sql
SELECT trim('   Spark SQL   ');
```

Output

```
Spark SQL
```

---

# 5. CONCAT()

```sql
SELECT concat('Data',' ','Engineer');
```

Output

```
Data Engineer
```

---

# 6. SUBSTRING()

Hive

```sql
SELECT substr('OpenAI',2,4);
```

Spark

```sql
SELECT substring('OpenAI',2,4);
```

Output

```
penA
```

---

# 7. INSTR()

Find position of substring.

```sql
SELECT instr('Data Engineering','Engineering');
```

Output

```
6
```

---

# 8. REPLACE()

```sql
SELECT replace('Hive SQL','Hive','Spark');
```

Output

```
Spark SQL
```

---

# 9. REGEXP_REPLACE()

Remove numbers.

```sql
SELECT regexp_replace('abc123xyz','[0-9]','');
```

Output

```
abcxyz
```

---

# 10. REGEXP_EXTRACT()

Extract first number.

```sql
SELECT regexp_extract('Order12345','\\d+',0);
```

Output

```
12345
```

---

# 11. SPLIT()

```sql
SELECT split('A,B,C,D',',');
```

Output

```
[A,B,C,D]
```

---

# 12. CONCAT_WS()

Join values using a separator.

```sql
SELECT concat_ws('-', '2026','07','23');
```

Output

```
2026-07-23
```

---

# 13. LPAD()

```sql
SELECT lpad('25',5,'0');
```

Output

```
00025
```

---

# 14. RPAD()

```sql
SELECT rpad('ABC',6,'*');
```

Output

```
ABC***
```

---

# 15. REVERSE()

```sql
SELECT reverse('Spark');
```

Output

```
krapS
```

---

# 16. REPEAT()

```sql
SELECT repeat('Hi',3);
```

Output

```
HiHiHi
```

---

# 17. INITCAP()

```sql
SELECT initcap('data engineer');
```

Output

```
Data Engineer
```

---

# 18. TRANSLATE()

Replace characters one-to-one.

```sql
SELECT translate('abc123','abc','xyz');
```

Output

```
xyz123
```

---

# 19. ASCII()

```sql
SELECT ascii('A');
```

Output

```
65
```

---

# 20. CHR()

```sql
SELECT chr(65);
```

Output

```
A
```

---

# 21. BASE64()

```sql
SELECT base64('OpenAI');
```

Output (example)

```
T3BlbkFJ
```

---

# 22. UNBASE64()

```sql
SELECT cast(unbase64('T3BlbkFJ') AS STRING);
```

Output

```
OpenAI
```

# Real-Time ETL Examples

## Remove Extra Spaces

```sql
SELECT trim(customer_name)
FROM customers;
```

---

## Convert Emails to Lowercase

```sql
SELECT lower(email)
FROM users;
```

---

## Standardize Product Codes

```sql
SELECT upper(product_code)
FROM products;
```

---

## Remove Special Characters

```sql
SELECT regexp_replace(phone,'[^0-9]','')
FROM customers;
```

---

## Extract Domain from Email

```sql
SELECT split(email,'@')[1]
FROM users;
```

Example:

```
john@gmail.com
```

Output

```
gmail.com
```

---

## Mask Credit Card

```sql
SELECT concat('************',substr(card_no,-4));
```

Output

```
************3456
```

---

## Format Employee IDs

```sql
SELECT lpad(emp_id,6,'0');
```

Output

```
000123
```

---

## Remove Multiple Spaces

```sql
SELECT regexp_replace(name,' +',' ');
```

---

## Build Full Name

```sql
SELECT concat(first_name,' ',last_name);
```

---

## Generate File Name

```sql
SELECT concat_ws('_','sales',current_date());
```

Output

```
sales_2026-07-23
```

# Most Asked Interview Functions

These are the functions you'll use most often in Hive and Spark SQL ETL jobs:

| Function                   | Use Case                            |
| -------------------------- | ----------------------------------- |
| `upper()`                  | Standardize text                    |
| `lower()`                  | Case normalization                  |
| `trim()`                   | Remove unwanted spaces              |
| `length()`                 | Validate string length              |
| `concat()`                 | Merge columns                       |
| `concat_ws()`              | Build delimited strings             |
| `substr()` / `substring()` | Extract parts of a string           |
| `instr()`                  | Search within strings               |
| `replace()`                | Replace literal text                |
| `regexp_replace()`         | Clean data with regular expressions |
| `regexp_extract()`         | Extract IDs, numbers, or patterns   |
| `split()`                  | Parse delimited data                |
| `lpad()` / `rpad()`        | Fixed-width formatting              |
| `initcap()`                | Proper-case names                   |
| `translate()`              | Character-by-character replacement  |

## Key Differences

* **Standard SQL** syntax varies by database vendor. For example, string concatenation is often `||`, while Hive and Spark primarily use `concat()`.
* **Hive and Spark SQL** are highly compatible for string functions, making it easy to move ETL logic between them.
* **Spark SQL** includes some additional functions (such as `soundex()` in many versions) and integrates tightly with DataFrame APIs.
* **Regular expression functions** (`regexp_replace`, `regexp_extract`) are among the most frequently used functions in modern data engineering because they simplify data cleansing and parsing.
