Hive provides several built-in functions for **date arithmetic**, allowing you to add/subtract days, calculate date differences, extract date parts, and manipulate timestamps. These are commonly used in ETL jobs, data warehousing, and reporting.

---

# 1. Current Date

Returns the current system date.

```sql
SELECT current_date;
```

**Output**

```
2026-07-23
```

---

# 2. Current Timestamp

Returns the current date and time.

```sql
SELECT current_timestamp;
```

**Output**

```
2026-07-23 14:35:20.123
```

---

# 3. Add Days (`date_add`)

Adds a specified number of days.

**Syntax**

```sql
date_add(date, number_of_days)
```

**Example**

```sql
SELECT date_add('2026-07-23', 10);
```

**Output**

```
2026-08-02
```

---

# 4. Subtract Days (`date_sub`)

Subtracts a specified number of days.

```sql
SELECT date_sub('2026-07-23', 7);
```

**Output**

```
2026-07-16
```

---

# 5. Difference Between Dates (`datediff`)

Returns the number of days between two dates.

**Syntax**

```sql
datediff(end_date, start_date)
```

**Example**

```sql
SELECT datediff('2026-07-23', '2026-07-01');
```

**Output**

```
22
```

# Real-Time Examples

## 1. Last 7 Days Data

```sql
SELECT *
FROM sales
WHERE order_date >= date_sub(current_date, 7);
```

---

## 2. Today's Orders

```sql
SELECT *
FROM orders
WHERE order_date = current_date;
```

---

## 3. Previous Month Data

```sql
SELECT *
FROM sales
WHERE month(order_date) = month(add_months(current_date, -1))
  AND year(order_date) = year(add_months(current_date, -1));
```

---

## 4. Customers Registered in Last 30 Days

```sql
SELECT customer_id
FROM customers
WHERE registration_date >= date_sub(current_date, 30);
```

---

## 5. Orders Older Than 90 Days

```sql
SELECT *
FROM orders
WHERE datediff(current_date, order_date) > 90;
```

---

## 6. Monthly Sales Report

```sql
SELECT
    year(order_date) AS yr,
    month(order_date) AS mon,
    SUM(amount) AS total_sales
FROM sales
GROUP BY year(order_date), month(order_date);
```

---

## 7. Daily Sales Report

```sql
SELECT
    order_date,
    SUM(amount) AS total_sales
FROM sales
GROUP BY order_date;
```

## Interview Tips

* `date_add()` and `date_sub()` work with **days only**. Use `add_months()` for month arithmetic.
* Use `datediff(end, start)` to calculate the number of days between two dates.
* `trunc()` works on **dates** (e.g., first day of the month or year), while `date_trunc()` works on **timestamps** (e.g., truncate to hour, day, month).
* For partition pruning in Spark SQL, prefer direct comparisons such as `order_date >= date_sub(current_date(), 7)` instead of applying functions to the partition column (e.g., `datediff(current_date(), order_date) <= 7`), as direct comparisons are generally more efficient.

| -------------------- | ---------------------------------------------------- | --------------------- | --------------------- | ------------------------------------------------------------ |

# Interview Summary

If you're interviewing for a Data Engineer role, a good way to explain the differences is:

* **Standard SQL** follows ANSI standards and is used in traditional relational databases such as PostgreSQL, Oracle, SQL Server, and MySQL (with vendor-specific extensions).
* **Hive SQL** extends SQL for Hadoop-based data warehousing. It introduces functions like `date_add()` and `date_sub()` because it doesn't rely heavily on ANSI `INTERVAL` syntax.
* **Spark SQL** is largely compatible with Hive SQL, but adds more modern functions such as `date_trunc()`, `months_between()`, `make_date()`, and `to_timestamp()`, making it more flexible for large-scale analytics and ETL.

> **Note:** "Normal SQL" isn't a single language—different database systems implement SQL differently. For example, Oracle uses `TRUNC()` for dates, PostgreSQL uses `DATE_TRUNC()`, SQL Server has its own `DATEADD()` and `DATEDIFF()`, and MySQL has `DATE_ADD()`. The "Standard SQL" column above refers to ANSI-style SQL, with notes where behavior is vendor-specific.
