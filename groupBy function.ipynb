
# PySpark groupBy() Function - Complete Guide

# What is groupBy()?

The `groupBy()` function groups rows that have the same values in one or more columns and then performs aggregate operations on each group.

Think of SQL:

```sql
SELECT Department,
       SUM(Salary)
FROM Employee
GROUP BY Department;
```

Equivalent PySpark:

```python
df.groupBy("Department").sum("Salary")
```

---

# Syntax

```python
df.groupBy(column_name).agg(function)
```

or

```python
df.groupBy("Department").sum("Salary")
```

or

```python
from pyspark.sql.functions import *

df.groupBy("Department").agg(
    sum("Salary"),
    avg("Salary"),
    max("Salary")
)
```

# 6. Multiple Aggregations

```python
from pyspark.sql.functions import *

df.groupBy("Department").agg(
    count("*").alias("Employees"),
    sum("Salary").alias("TotalSalary"),
    avg("Salary").alias("AverageSalary"),
    max("Salary").alias("HighestSalary"),
    min("Salary").alias("LowestSalary")
).show()
```

Output

|Department|Employees|TotalSalary|AverageSalary|HighestSalary|LowestSalary|
|----------|----------|-----------|-------------|--------------|-------------|
|IT|3|18000|6000|7000|5000|
|HR|2|8500|4250|4500|4000|
|Sales|2|15500|7750|8000|7500|

# groupBy() vs Window Function

Suppose we have:

|Employee|Department|Salary|
|----------|----------|------|
|John|IT|5000|
|David|IT|7000|
|Alex|IT|6000|

Using groupBy

```python
df.groupBy("Department").avg("Salary")
```

Output

|Department|Average|
|----------|--------|
|IT|6000|

Employee rows are gone.

---

Using Window Function

```python
windowSpec = Window.partitionBy("Department")

df.withColumn(
    "AverageSalary",
    avg("Salary").over(windowSpec)
)
```

Output

|Employee|Department|Salary|Average|
|----------|----------|------|-------|
|John|IT|5000|6000|
|David|IT|7000|6000|
|Alex|IT|6000|6000|

Rows are preserved.

# Performance Tips

✔ Filter data before `groupBy()` whenever possible.

```python
df.filter(col("Salary") > 5000) \
  .groupBy("Department") \
  .sum("Salary")
```

✔ Avoid unnecessary grouping on high-cardinality columns (e.g., UUIDs).

✔ Select only the columns you need before aggregation.

```python
df.select("Department", "Salary") \
  .groupBy("Department") \
  .sum("Salary")
```

✔ Be aware that `groupBy()` causes a **shuffle**, which is one of the most expensive Spark operations.

✔ Repartition on grouping keys if you're repeatedly aggregating large datasets by the same columns.

# Common Interview Questions

### Q1. Does `groupBy()` cause a shuffle?

**Answer:** Yes. Spark redistributes data across partitions so that rows with the same grouping key end up together.

---

### Q2. Can we use multiple aggregation functions together?

```python
df.groupBy("Department").agg(
    sum("Salary"),
    avg("Salary"),
    max("Salary"),
    min("Salary"),
    count("*")
)
```

---

### Q3. What is the difference between `groupBy()` and `reduceByKey()`?

- `groupBy()` works on **DataFrames** and is optimized using Spark SQL/Catalyst.
- `reduceByKey()` works on **RDDs**.

---

### Q4. When should we use `groupBy()` vs a Window Function?

- Use **groupBy()** when you need one aggregated row per group.
- Use **Window Functions** when you need aggregated values while keeping every original row.
