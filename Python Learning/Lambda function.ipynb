# 1. What is a Lambda Function?

A **lambda function** is a small anonymous function in Python that is created using the `lambda` keyword.

It is mainly used when:

- A small function is required temporarily
- Passing a function as an argument
- Performing quick transformations
- Sorting and filtering data

---

## Normal Function Example

```python
def square(x):
    return x * x


print(square(5))
```

Output:

```
25
```

---

## Lambda Function Example

```python
square = lambda x: x * x

print(square(5))
```

Output:

```
25
```

---

# 2. Lambda Syntax

```python
lambda arguments : expression
```

Example:

```python
lambda x: x + 10
```

Explanation:

```
Input  : x
Process: x + 10
Output : Result
```

---

# 3. Lambda Rules

Lambda functions:

✅ Can accept multiple arguments  
✅ Return one expression automatically  
✅ Used with map(), filter(), sorted()  
✅ Useful for short operations  

Lambda limitations:

❌ Cannot contain multiple statements  
❌ Cannot replace complex functions  
❌ Harder to debug for large logic

---

# 4. Lambda vs Normal Function

## Normal Function

```python
def add(a,b):
    return a+b
```

## Lambda Function

```python
add = lambda a,b:a+b
```

---

## Comparison

| Normal Function | Lambda Function |
|---|---|
| Uses `def` | Uses `lambda` |
| Has function name | Usually anonymous |
| Supports multiple statements | Single expression |
| Better for large logic | Better for small logic |
| Reusable | Usually temporary |

# 7. Lambda with map()

`map()` applies a function to every item in a list.

Syntax:

```python
map(function, iterable)
```

---

## Example

```python
numbers = [
    1,
    2,
    3,
    4,
    5
]


squares = list(
    map(
        lambda x:x*x,
        numbers
    )
)


print(squares)
```

Output:

```
[1,4,9,16,25]
```

---

# 8. Lambda with filter()

`filter()` selects items based on a condition.

Syntax:

```python
filter(function, iterable)
```

---

## Example

Find even numbers:

```python
numbers = [
    1,
    2,
    3,
    4,
    5,
    6
]


even = list(
    filter(
        lambda x:x%2==0,
        numbers
    )
)


print(even)
```

Output:

```
[2,4,6]
```

# 11. Lambda with Conditional Logic

Lambda supports if-else.

Syntax:

```python

```

---

Example:

```python

print(check_age(20))

print(check_age(15))
```

Output:

```

Minor
```

# 13. Real-Time Project

## Employee Ranking System

```python
employees = [

    {
        "name":"John",
        "department":"IT",
        "salary":60000
    },

    {
        "name":"Alice",
        "department":"HR",
        "salary":70000
    },

    {
        "name":"Bob",
        "department":"IT",
        "salary":50000
    }

]


ranking = sorted(
    employees,
    key=lambda x:x["salary"],
    reverse=True
)


for rank,employee in enumerate(ranking,start=1):

    print(
        rank,
        employee["name"],
        employee["salary"]
    )
```

Output:

```
1 Alice 70000
2 John 60000
3 Bob 50000
```

# 14. Lambda Interview Questions

## Q1. Can lambda accept multiple arguments?

Yes.

Example:

```python
add = lambda a,b,c:a+b+c

print(add(1,2,3))
```

Output:

```
6
```

---

## Q2. Can lambda contain multiple statements?

No.

Wrong:

```python
lambda x:
    y=x+1
    return y
```

Use normal function instead.

---

## Q3. Why use lambda with sorted()?

Because sorting often requires temporary logic.

Example:

```python
employees.sort(
    key=lambda x:x["salary"]
)
```
