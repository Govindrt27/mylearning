# Mini Project – Email Validator

```python
email = input("Enter Email: ").strip()

if (
    "@" in email
    and email.endswith(".com")
    and " " not in email
):
    username, _, domain = email.partition("@")

    print("Valid Email")
    print("Username:", username)
    print("Domain:", domain)

else:
    print("Invalid Email")
```

Methods used

- strip()
- endswith()
- partition()
- in operator
