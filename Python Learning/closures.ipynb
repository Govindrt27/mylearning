# 1. What is a Closure?

A **closure** is a function that remembers the variables from its outer function **even after the outer function has finished executing**.

In simple words,

> **A closure is a function with memory.**

---

# Real-Life Analogy

Imagine your mother packs a lunch box before you leave for school.

The lunch box contains:

- Sandwich
- Juice
- Chocolate

Even after leaving home, you still have everything because the lunch box **remembers** what was packed.

A closure behaves in the same way.

The inner function remembers the variables that were created in the outer function.

---

# 2. Why Do We Need Closures?

- Create customized functions

# 3. Closure vs Normal Function

## Normal Function

```python
def greet():
    name = "John"

    print("Hello", name)


greet()
```

Output

```
Hello John
```

After the function finishes, the variable `name` is destroyed.

---

## Closure

```python
def outer():

    name = "John"

    def inner():
        print("Hello", name)

    return inner


greeting = outer()

greeting()
```

Output

```
Hello John
```

Although `outer()` has finished executing, the `inner()` function still remembers `name`.

---

# 4. How a Closure Works

Step 1

```python
def outer():

    name = "John"

    def inner():
        print(name)

    return inner
```

---

Step 2

```python
greeting = outer()
```

Python performs:

```
Create variable

name = "John"

↓

Create inner()

↓

Return inner()
```

---

Step 3

Even after `outer()` finishes,

```
name = "John"
```

is still remembered because `inner()` needs it.

This remembered variable is called a **closure**.

---

# Visual Representation

```
outer()

+----------------------+
| name = "John"        |
|                      |
| inner()              |
+----------------------+
          |
          |
          V
Return inner()
          |
          |
          V
greeting()

Still remembers

name = "John"
```

# 5. First Closure Example

```python
def multiplier(number):

    def multiply(value):
        return value * number

    return multiply
```

Create a function:

```python
double = multiplier(2)

print(double(10))
```

Output

```
20
```

Explanation

The function remembers:

```
number = 2
```

Later,

```python
double(10)
```

actually performs:

```
10 × 2
```

---

Another function

```python
triple = multiplier(3)

print(triple(10))
```

Output

```
30
```

Here the remembered value is

```
number = 3
```

# 6. Understanding the Memory

```python
def outer():

    x = 100

    def inner():
        print(x)

    return inner
```

Calling

```python
obj = outer()
```

creates

```
x = 100
```

Normally this variable would disappear after `outer()` finishes.

However,

because `inner()` still needs `x`,

Python keeps it alive.

This stored variable is the **closure**.

## Example 3: Logger




The logger remembers the module name.

# 8. Inspecting a Closure

Python stores remembered variables internally.

Example

```python
def outer():

    x = 10

    def inner():
        print(x)

    return inner


obj = outer()

print(obj.__closure__)
```

Output (similar to)

```
(<cell at 0x...: int object>,)
```

This shows that Python has stored the remembered variable.

# 10. Closures vs Classes

Closure

```python
def multiplier(n):

    def multiply(x):
        return x * n

    return multiply
```

Class

```python
class Multiplier:

    def __init__(self, n):
        self.n = n

    def multiply(self, x):
        return x * self.n
```

Comparison

| Closure | Class |
|---------|-------|
| Simple | More powerful |
| Good for small state | Good for complex state |
| Less code | More structure |
| Function-based | Object-oriented |

# Easy Rule to Remember

A function is a **closure** if:

- It is defined inside another function.
- It uses variables from the outer function.
- The outer function returns the inner function.
- The inner function remembers those variables after the outer function has finished.

Simply remember:

> **Closure = Function + Memory**
