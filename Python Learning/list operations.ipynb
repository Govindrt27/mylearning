# 1. Introduction to Lists

A **List** is one of Python's most commonly used built-in data structures. It stores multiple values in a single variable and preserves the order in which items are added.

Lists are:

- Ordered
- Mutable (modifiable)
- Allow duplicate values
- Can store different data types
- Indexed starting from 0

## Syntax

```python
my_list = [1, 2, 3]
```

---

## Real-Time Use Cases

Lists are widely used in real-world applications.

Examples:

- Shopping cart items
- Student records
- Product catalogs
- Employee names
- Customer orders
- Chat messages
- Recent notifications
- Image galleries

Example

```python
shopping_cart = [
    "Laptop",
    "Mouse",
    "Keyboard"
]

print(shopping_cart)
```

Output

```
['Laptop', 'Mouse', 'Keyboard']
```

# 1. append()

The `append()` method adds **one item** to the **end** of a list.

## Syntax

```python
list.append(item)
```

---

## Example 1

```python
fruits = ["Apple", "Banana"]

fruits.append("Orange")

print(fruits)
```

Output

```
['Apple', 'Banana', 'Orange']
```

---

## Example 2 - Numbers

```python
numbers = [10, 20, 30]

numbers.append(40)

print(numbers)
```

Output

```
[10, 20, 30, 40]
```

---

## Example 3 - Append a List

```python
numbers = [1, 2, 3]

numbers.append([4, 5])

print(numbers)
```

Output

```
[1, 2, 3, [4, 5]]
```

Notice that the **entire list becomes a single element**.

# 2. extend()

The `extend()` method adds **multiple elements** from another iterable.

## Syntax

```python
list.extend(iterable)
```

---

## Example 1

```python
list1 = [1, 2]

list2 = [3, 4]

list1.extend(list2)

print(list1)
```

Output

```
[1, 2, 3, 4]
```

---

## Example 2 - Extend with Tuple

```python
numbers = [10, 20]

numbers.extend((30, 40))

print(numbers)
```

Output

```
[10, 20, 30, 40]
```

---

## Example 3 - Extend with String

```python
letters = ["A"]

letters.extend("BCD")

print(letters)
```

Output

```
['A', 'B', 'C', 'D']
```

---

## Difference Between append() and extend()

```python
a = [1, 2]

a.append([3, 4])

print(a)
```

Output

```
[1, 2, [3, 4]]
```

---

```python
a = [1, 2]

a.extend([3, 4])

print(a)
```

Output

```
[1, 2, 3, 4]
```

# 5. pop()

The `pop()` method removes an item by index and returns it.

If no index is provided, it removes the last element.

---

## Syntax

```python
list.pop(index)
```

or

```python
list.pop()
```

---

## Example 1

```python
numbers = [10, 20, 30]

removed = numbers.pop()

print(removed)

print(numbers)
```

Output

```
30

[10, 20]
```

---

## Example 2

```python
numbers = [10, 20, 30]

removed = numbers.pop(1)

print(removed)

print(numbers)
```

Output

```
20

[10, 30]
```

---

## Example 3

```python
stack = []

stack.append("A")
stack.append("B")
stack.append("C")

print(stack.pop())

print(stack)
```

Output

```
C

['A', 'B']
```

---

## Real-Time Use Cases

### Browser History

```python
history = [
    "Google",
    "YouTube",
    "GitHub"
]

last_page = history.pop()

print(last_page)

print(history)
```

---

### Undo Feature

```python
actions = [
    "Draw Circle",
    "Draw Square",
    "Draw Line"
]

undo = actions.pop()

print("Undo:", undo)

print(actions)
```

---

### Stack Implementation

```python
stack = []

stack.append("Login")

stack.append("Dashboard")

stack.append("Profile")

while stack:
    print(stack.pop())
```

Output

```
Profile
Dashboard
Login
```

---

## Time Complexity

| Operation | Complexity |
|------------|------------|
| pop() (last) | O(1) |
| pop(index) | O(n) |

# 3. index()

The `index()` method returns the position of the first occurrence of a value.

## Syntax

```python
list.index(value)
```

---

## Example

```python
colors = [
    "Red",
    "Green",
    "Blue"
]

position = colors.index("Green")

print(position)
```

Output

```
1
```

---

## Duplicate Values

```python
numbers = [
    10,
    20,
    10,
    30
]

print(numbers.index(10))
```

Output

```
0
```

It returns the first match only.

---


Output

```

```

# 7. Shallow Copy vs Deep Copy

This is an important interview topic.

---

# Shallow Copy

A shallow copy creates a new outer list but keeps references to nested objects.

Example:

```python
original = [
    [1,2],
    [3,4]
]

copy_list = original.copy()

copy_list[0][0] = 99

print(original)

print(copy_list)
```

Output

```
[[99,2],[3,4]]

[[99,2],[3,4]]
```

Both changed.

---

# Deep Copy

Deep copy creates completely independent objects.

Use:

```python
import copy
```

Example:

```python
import copy

original = [
    [1,2],
    [3,4]
]

deep = copy.deepcopy(original)

deep[0][0] = 99

print(original)

print(deep)
```

Output

```
[[1,2],[3,4]]

[[99,2],[3,4]]
```
