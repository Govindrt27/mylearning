**Time Complexity** is a way to measure how the execution time of an algorithm grows as the input size (`n`) increases.

Think of it as:

> "If the amount of data increases, how many operations will my code perform?"

***

## Common Time Complexities

### O(1) - Constant Time

The number of operations never changes.

```python
students = ["Alice", "Bob", "Charlie"]

print(students[1])
```

No matter whether there are 3 students or 3 million students, accessing an index takes the same amount of time.

***

### O(log n) - Logarithmic Time

The data is cut in half at every step.

Example: Binary Search

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

Example:

```text
16 elements
↓
8 elements
↓
4 elements
↓
2 elements
↓
1 element
```

Only 4 searches are needed for 16 elements.

***

### O(n) - Linear Time

The algorithm processes each element once.

```python
def linear_search(arr, target):
    for item in arr:
        if item == target:
            return True
    return False
```

Example:

```text
1
2
3
4
5
6
7
```

To find 7, it may need to check all 7 elements.

***

### O(n log n) - Linearithmic Time

Common in efficient sorting algorithms.

```python
sorted_numbers = sorted(numbers)
```

Examples:

* Merge Sort
* Heap Sort
* Quick Sort (average case)

***

### O(n²) - Quadratic Time

Nested loops.

```python
for i in range(n):
    for j in range(n):
        print(i, j)
```

If:

```text
n = 10  → 100 operations
n = 100 → 10,000 operations
```

Growth becomes much faster.

***

## Visual Comparison

Suppose:


## Interview Tip

When asked for time complexity:

1. Count loops.
2. Check if loops are nested.
3. Check if the search space is halved (`log n`).
4. Ignore constants.

Example:

```python
for i in range(n):
    print(i)
```

Complexity:

```text
n operations
=> O(n)
```

Example:

```python
for i in range(n):
    for j in range(n):
        print(i, j)
```

Complexity:

```text
n × n
=> O(n²)
```

Example:

```python
while n > 1:
    n = n // 2
```

Complexity:

```text
n → n/2 → n/4 → n/8 ...
=> O(log n)
```

### Rule of Thumb

* **One loop over n items** → **O(n)**
* **Nested loops over n items** → **O(n²)**
* **Halving the data each iteration** → **O(log n)**
* **Direct access by index/key** → **O(1)**


````python
import time

students = [
    {"id": 101 + i, "name": f"Student_{101 + i}"} for i in range(1000000)
]

def find_student_binary(students, student_id):
    left = 0
    right = len(students) - 1

    while left <= right:
        mid = (left + right) // 2
        current_id = students[mid]["id"]

        if current_id == student_id:
            return students[mid]
        elif current_id < student_id:
            left = mid + 1
        else:
            right = mid - 1

    return None

start_time = time.time()
print(find_student_binary(students, 195769))
end_time = time.time()
print(f"Time taken: {(end_time - start_time) * 1000:.2f} ms")
````

````python
students = [
    {"id": 101 + i, "name": f"Student_{101 + i}"} for i in range(1000000)
]

def find_student_linear(students, student_id):
    for student in students:
        if student["id"] == student_id:
            return student
    return None

start_time = time.time()
print(find_student_linear(students, 195769))
end_time = time.time()
print(f"Time taken: {(end_time - start_time) * 1000:.2f} ms")
````

## Why O(log n)?

Notice that in each iteration we eliminate **half** of the remaining records.

```text
7 elements
   ↓
3 elements
   ↓
1 element
```

For a million sorted records:

```text
1,000,000
↓
500,000
↓
250,000
↓
125,000
...
↓
≈20 steps
```

That's why Binary Search runs in **O(log n)** time instead of **O(n)**.


````python
students = [
    {"id": 101, "name": "Alice"},
    {"id": 102, "name": "Bob"},
    {"id": 103, "name": "Charlie"},
    {"id": 104, "name": "David"},
    {"id": 105, "name": "Emma"},
    {"id": 106, "name": "Frank"},
]

def find_student_binary(students, student_id):
    left = 0
    right = len(students) - 1

    while left <= right:
        mid = (left + right) // 2
        current_id = students[mid]["id"]

        if current_id == student_id:
            return students[mid]
        elif current_id < student_id:
            left = mid + 1
        else:
            right = mid - 1

    return None

print(find_student_binary(students, 106))
````
