# 1. What is a Decorator?

A **decorator** is a function that **adds extra functionality to another function without modifying its original code**.

In simple words:

> **Decorator = Wrap an existing function with additional behavior**

Think of it like wrapping a gift.

- 🎁 Gift → Original Function
- 🎀 Gift Wrapper → Decorator

The gift remains the same, but the wrapper adds something extra.

---

# Real-Life Analogy

Imagine you order a pizza.

The chef prepares the pizza.

Before delivery, the restaurant:

- Packs it into a box
- Adds sauce packets
- Adds tissues
- Seals the package

The pizza itself never changes.

The restaurant simply **wraps additional functionality around it.**

A Python decorator works exactly the same way.

```
Original Function
        │
        ▼
Decorator
(Add Extra Work)
        │
        ▼
Final Function
```

---

# 2. Why Do We Need Decorators?

Suppose your application has **100 functions**.

Before every function executes, you want to:

- Check user authentication
- Log function calls
- Measure execution time
- Handle exceptions
- Validate permissions

Without decorators, you would repeat the same code in every function.

With decorators, you write that code **once** and reuse it everywhere.

# 3. Simple Function Example

```python
def greet():
    print("Hello World")


greet()
```

Output

```
Hello World
```

Suppose we want the output to become

```
Starting...

Hello World

Finished...
```

We could modify the function.

```python
def greet():

    print("Starting...")

    print("Hello World")

    print("Finished...")
```

But imagine doing this for 500 functions.

Instead, decorators solve this problem.

---

# 4. First Decorator

```python
def decorator(func):

    def wrapper():

        print("Starting...")

        func()

        print("Finished...")

    return wrapper
```

Original function

```python
def greet():

    print("Hello World")
```

Decorate the function

```python
greet = decorator(greet)

greet()
```

Output

```
Starting...

Hello World

Finished...
```

Notice that the original function never changed.

---

# 5. Decorator Syntax (@)

Python provides a cleaner syntax.

Instead of

```python
greet = decorator(greet)
```

we write

```python
@decorator
def greet():

    print("Hello World")
```

Calling

```python
greet()
```

produces the same result.

---

# What Happens Internally?

Python converts

```python
@decorator
def greet():

    print("Hello")
```

into

```python
def greet():

    print("Hello")


greet = decorator(greet)
```

The `@decorator` syntax is simply a shortcut.

# 6. Decorators with Function Arguments

Functions that accept parameters can also be decorated.

```python
def logger(func):

    def wrapper(name):

        print("Function Started")

        func(name)

        print("Function Finished")

    return wrapper
```

Use it

```python
@logger
def greet(name):

    print("Hello", name)


greet("John")
```

Output

```
Function Started

Hello John

Function Finished
```

# 7. Closures and Decorators

Every decorator uses a **closure**.

A closure is a function that remembers variables from its outer function.

In a decorator,

- `func` is remembered by the inner wrapper function.
- The wrapper adds extra functionality.
- The wrapper is returned.

Therefore,

```
Closure
      +
Wrapper Function
      =
Decorator
```

# 9. Real-Time Example 2 - Authentication

Only logged-in users can access profile information.

```python
logged_in = True


def login_required(func):

    def wrapper():

        if logged_in:

            func()

        else:

            print("Access Denied")

    return wrapper
```

Use it

```python
@login_required
def profile():

    print("Profile Details")


profile()
```

Output

```
Profile Details
```

If

```python
logged_in = False
```

Output

```
Access Denied
```

# 10. Real-Time Example 3 - Measure Execution Time

```python
import time


def timer(func):

    def wrapper():

        start = time.time()

        func()

        end = time.time()

        print("Execution Time:", end - start)

    return wrapper
```

Use it

```python
@timer
def process_data():

    time.sleep(2)

    print("Processing Complete")


process_data()
```

Output

```
Processing Complete

Execution Time: 2.00 Seconds
```

Applications

- ETL pipelines
- Machine Learning
- Data Engineering
- API Performance

# 11. Real-Time Example 4 - Exception Handling

```python
def safe(func):

    def wrapper():

        try:

            func()

        except Exception as error:

            print("Error:", error)

    return wrapper
```

Use it

```python
@safe
def divide():

    print(10 / 0)


divide()
```

Output

```
Error: division by zero
```

# 13. Real-Time Example 6 - Retry Failed API Calls

```python
def retry(func):

    def wrapper():

        for attempt in range(3):

            try:

                return func()

            except Exception:

                print("Retrying...")

    return wrapper

```

# 14. Built-in Decorators

Python provides several built-in decorators.

## @staticmethod

```python
class Calculator:

    @staticmethod
    def add(a, b):

        return a + b
```

---

## @classmethod

```python
class Employee:

    company = "ABC"

    @classmethod
    def show_company(cls):

        print(cls.company)
```

---

## @property

```python
class Employee:

    @property
    def salary(self):

        return 50000
```

---

## @functools.lru_cache

Caches expensive function calls.

```python
from functools import lru_cache


@lru_cache
def fibonacci(n):

    if n < 2:

        return n

    return fibonacci(n - 1) + fibonacci(n - 2)
```

Applications

- Dynamic Programming
- Recursive Algorithms
- Performance Optimization

# 16. Decorators vs Closures

| Closure | Decorator |
|----------|-----------|
| Remembers variables | Adds extra functionality |
| Stores state | Wraps another function |
| Returns inner function | Returns wrapper function |
| Used internally by decorators | Built using closures |

# 17. Common Interview Questions

## What is a decorator?

A decorator is a function that wraps another function to add extra functionality without modifying the original function.

---

## Why use decorators?

To reuse common logic like:

- Logging
- Authentication
- Authorization
- Timing
- Validation
- Exception Handling

---

## Can decorators accept arguments?

Yes.

Both the decorated function and the decorator itself can accept arguments.

---

## Are decorators built using closures?

Yes.

Decorators are implemented using closures because the wrapper function remembers the original function.

---

## What is the benefit of the @ symbol?

It is simply syntactic sugar.

Python converts

```python
@decorator
def func():
    pass
```

into

```python
func = decorator(func)
```

A **decorator** is a powerful Python feature that allows you to add functionality to an existing function **without changing its original implementation**. Decorators are widely used in real-world applications for logging, authentication, authorization, timing, validation, exception handling, caching, and performance monitoring.
