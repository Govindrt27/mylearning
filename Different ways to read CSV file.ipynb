```python
import pandas as pd
import numpy as np

np.random.seed(42)
num_rows = 100

countries = ['USA', 'Canada', 'UK', 'Germany', 'France']
product_names = ['Widget', 'Gadget', 'Doodad', 'Thingamajig', 'Contraption']

data = {
    'product_id': np.random.randint(1000, 1100, num_rows),
    'product_name': np.random.choice(product_names, num_rows),
    'country': np.random.choice(countries, num_rows),
    'sales': np.random.randint(10, 1000, num_rows),
    'date': pd.to_datetime(np.random.choice(pd.date_range('2025-01-01', '2025-12-31'), num_rows)),
}
df = pd.DataFrame(data)
df['year'] = df['date'].dt.year

df.to_csv('/dbfs/tmp/dummy_product_sales.csv', index=False)
```
