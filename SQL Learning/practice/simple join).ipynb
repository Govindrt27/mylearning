```python
from pyspark.sql import Row

# Create DataFrames for table1 and table2
table1 = spark.createDataFrame([Row(value=1), Row(value=1), Row(value=1), Row(value=2), Row(value=None), Row(value=3), Row(value=3)])
table2 = spark.createDataFrame([Row(value=1), Row(value=1), Row(value=None), Row(value=4), Row(value=4)])

table1.createOrReplaceTempView("table1")
table2.createOrReplaceTempView("table2")

# Inner Join
inner_join = spark.sql("SELECT t1.value as t1_value, t2.value as t2_value FROM table1 t1 INNER JOIN table2 t2 ON t1.value = t2.value")
display(inner_join)

# Left Join
left_join = spark.sql("SELECT t1.value as t1_value, t2.value as t2_value FROM table1 t1 LEFT JOIN table2 t2 ON t1.value = t2.value")
display(left_join)

# Right Join
right_join = spark.sql("SELECT t1.value as t1_value, t2.value as t2_value FROM table1 t1 RIGHT JOIN table2 t2 ON t1.value = t2.value")
display(right_join)

# Full Outer Join
full_outer_join = spark.sql("SELECT t1.value as t1_value, t2.value as t2_value FROM table1 t1 FULL OUTER JOIN table2 t2 ON t1.value = t2.value")
display(full_outer_join)

# Cross Join
cross_join = spark.sql("SELECT t1.value as t1_value, t2.value as t2_value FROM table1 t1 CROSS JOIN table2 t2")
display(cross_join)
```
