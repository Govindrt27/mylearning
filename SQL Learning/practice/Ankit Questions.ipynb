
### Write a query to display records which have 3 or more consecutive rows with amount of people are more than 100 each day

```python
%sql
create table stadium (
id int,
visit_date date,
no_of_people int
);

insert into stadium
values (1,'2017-07-01',10)
,(2,'2017-07-02',109)
,(3,'2017-07-03',150)
,(4,'2017-07-04',99)
,(5,'2017-07-05',145)
,(6,'2017-07-06',1455)
,(7,'2017-07-07',199)
,(8,'2017-07-08',188);
```

```python
%sql
select * from stadium
```

```python
%sql

select * from stadium where no_of_people> 100
```

```python
%sql

select *, row_number() over(order by visit_date asc) as rnk from stadium
where no_of_people>100
```

```python
%sql
select *, row_number() over(order by visit_date asc) as rnk,
id-row_number() over(order by visit_date asc) as grp
from stadium
where no_of_people>100
```

```python
%sql

with grped_data (
    select *, row_number() over(order by visit_date asc) as rnk,
id-row_number() over(order by visit_date asc) as grp
from stadium
where no_of_people>100
)

select grp,count(grp) from grped_data group by grp having count(grp)>3
```

```python
%sql

with grped_data (
    select *, row_number() over(order by visit_date asc) as rnk,
id-row_number() over(order by visit_date asc) as grp
from stadium
where no_of_people>100
)

select * from grped_data WHERE grp in 
(select grp from (select grp,count(grp) from grped_data group by grp having count(grp)>3))
```

```python
%sql
with tbl as (
    select case when city="Mumbai" then name else null end as Mumbai,
    case when city="Bangalore" then name else null end as Bangalore,
    case when city="Delhi" then name else null end as Delhi
    from players_location
)

select max(distinct Mumbai),Bangalore,Delhi from tbl group by Mumbai,Bangalore,Delhi order by Mumbai,Bangalore,Delhi desc
```

```python
%sql
with tbl as (select * ,row_number() over(partition by city order by name) as rnk
from players_location)

select rnk,max(case when city="Mumbai" then name else null end) as Mumbai,
max(case when city="Bangalore" then name else null end) as Bangalore,
max(case when city="Delhi" then name else null end) as Delhi
from tbl 
group by rnk
```
