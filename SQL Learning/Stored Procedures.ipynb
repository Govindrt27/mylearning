## What is a Stored Procedure in SQL?

A **Stored Procedure** is a precompiled collection of one or more SQL statements stored in the database. It can be executed whenever needed to perform a specific task, such as inserting, updating, deleting, or retrieving data.

Think of a stored procedure as a **reusable function for the database**.

### Why Use Stored Procedures?

* **Reusability** – Write once, use many times.
* **Better Performance** – Execution plans are cached by the database.
* **Security** – Users can be granted access to execute procedures without direct access to tables.
* **Maintainability** – Business logic is centralized in the database.
* **Reduced Network Traffic** – Execute multiple SQL statements using a single call.

***

## Example 1: Simple Stored Procedure

```sql
CREATE PROCEDURE GetEmployees
AS
BEGIN
    SELECT *
    FROM Employees;
END;
```

Execute:

```sql
EXEC GetEmployees;
```

***

## Example 2: Stored Procedure with Parameters

```sql
CREATE PROCEDURE GetEmployeeById
    @EmployeeID INT
AS
BEGIN
    SELECT *
    FROM Employees
    WHERE EmployeeID = @EmployeeID;
END;
```

Execute:

```sql
EXEC GetEmployeeById @EmployeeID = 101;
```

***

## Example 3: Insert Data Using Stored Procedure

```sql
CREATE PROCEDURE InsertEmployee
    @EmployeeID INT,
    @EmployeeName VARCHAR(100),
    @Salary DECIMAL(10,2)
AS
BEGIN
    INSERT INTO Employees
    VALUES (@EmployeeID, @EmployeeName, @Salary);
END;
```

Execute:

```sql
EXEC InsertEmployee
    @EmployeeID = 101,
    @EmployeeName = 'John',
    @Salary = 50000;
```

***

## Components of a Stored Procedure

```sql
CREATE PROCEDURE ProcedureName
    @Parameter1 DataType,
    @Parameter2 DataType
AS
BEGIN
    -- SQL Statements
END;
```

### Key Elements

* **CREATE PROCEDURE** – Creates the procedure.
* **Parameters** – Input values passed during execution.
* **BEGIN...END** – Contains SQL logic.
* **EXEC** – Executes the procedure.

***

## Stored Procedure vs Function

| Stored Procedure                   | Function                        |
| ---------------------------------- | ------------------------------- |
| Can return multiple result sets    | Returns a single value/table    |
| Can perform INSERT, UPDATE, DELETE | Primarily used for calculations |
| Can use transactions               | Limited transaction support     |
| Executed using `EXEC`              | Called within SQL statements    |

Example Function:

```sql
SELECT dbo.CalculateBonus(50000);
```

Example Procedure:

```sql
EXEC UpdateEmployeeSalary;
```

## Interview Questions

### Q1. What is a Stored Procedure?

**Answer:**  
A stored procedure is a precompiled set of SQL statements stored in a database and executed as a single unit to perform specific business operations.

### Q2. What are the advantages of Stored Procedures?

**Answer:**

* Improved performance
* Code reusability
* Better security
* Centralized business logic
* Easier maintenance

### Q3. Can a Stored Procedure return values?

**Answer:**  
Yes. It can return:

* Result sets
* Output parameters
* Return status codes

Example:

```sql
CREATE PROCEDURE GetEmployeeCount
    @Count INT OUTPUT
AS
BEGIN
    SELECT @Count = COUNT(*)
    FROM Employees;
END;
```

### Q4. What is the difference between EXEC and EXECUTE?

**Answer:**  
No difference. `EXEC` is the shortened form of `EXECUTE`.

```sql
EXEC GetEmployees;

EXECUTE GetEmployees;
```

### Q5. Can we call one Stored Procedure from another?

**Answer:**  
Yes.

```sql
CREATE PROCEDURE MainProcedure
AS
BEGIN
    EXEC GetEmployees;
END;
```

## Real-Time ADF Usage

In **Azure Data Factory (ADF)**, Stored Procedures are commonly used to:

* Load audit tables
* Update control tables
* Record pipeline execution status
* Implement incremental load logic
* Perform post-load validations

Example ADF workflow:

```text
Copy Activity
       ↓
Stored Procedure Activity
       ↓
Update Audit Table
```

### Example

```sql
CREATE PROCEDURE UpdateLoadStatus
    @PipelineName VARCHAR(100),
    @Status VARCHAR(20)
AS
BEGIN
    INSERT INTO AuditTable
    VALUES (@PipelineName, @Status, GETDATE());
END;
```

ADF can call this procedure using the **Stored Procedure Activity**.
