```python
#3. Handle Nested Arrays using explode()If your JSON contains an array of objects, Spark reads it as an ArrayType. To parse it, use the explode() function to transform each element of the array into its own row

from pyspark.sql.functions import explode, col

# Step 1: Explode the array column into rows of individual structs
exploded_df = df.select(col("order_id"), explode(col("items")).alias("item"))

# Step 2: Extract fields from the exploded struct using dot notation
final_df = exploded_df.select(
    col("order_id"),
    col("item.item_name").alias("item_name"),
    col("item.price").alias("price")
)
final_df.show()


#2. Parse Nested Objects (StructType) using Dot NotationWhen a JSON object contains nested objects, Spark reads them as a StructType. You can access inner fields directly using dot notation (parent.child) inside a select statement.

from pyspark.sql.functions import col

# Extract nested fields and assign new column names
flattened_df = df.select(
    col("id"),
    col("user.name").alias("user_name"),
    col("user.city").alias("user_city")
)
flattened_df.show()


#
#4. Parse JSON Strings inside Columns via from_jsonIf your DataFrame already contains a raw JSON string inside a standard text column, use from_json() to parse it dynamically. This requires explicitly defining a Spark schema.

from pyspark.sql.functions import from_json, col
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# 1. Define the structural schema of the embedded JSON string
json_schema = StructType([
    StructField("age", IntegerType(), True),
    StructField("gender", StringType(), True)
])

# 2. Parse the string column and un-nest the fields
parsed_df = df.withColumn("parsed_meta", from_json(col("raw_string_column"), json_schema)) \
              .select("id", "parsed_meta.age", "parsed_meta.gender")

parsed_df.show()
```

```python
from pyspark.sql.functions import explode, col
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, ArrayType

json_data = [
    {
        "customer_id": 101,
        "status": "active",
        "name": {"first": "John", "last": "Doe"},
        "address": {"street": "123 Elm St", "city": "Springfield", "zip": "62704"},
        "orders": [
            {
                "order_id": "A001",
                "amount": 250,
                "items": [
                    {"product": "Laptop", "qty": 1},
                    {"product": "Mouse", "qty": 2}
                ]
            },
            {
                "order_id": "A002",
                "amount": 150,
                "items": [
                    {"product": "Keyboard", "qty": 1}
                ]
            }
        ]
    },
    {
        "customer_id": 102,
        "status": "active",
        "name": {"first": "John1", "last": "Doe1"},
        "address": {"street": "123 Elm St", "city": "Springfield", "zip": "62704"},
        "orders": [
            {
                "order_id": "A003",
                "amount": 250,
                "items": [
                    {"product": "Laptop", "qty": 1},
                    {"product": "Mouse", "qty": 2}
                ]
            },
            {
                "order_id": "A004",
                "amount": 150,
                "items": [
                    {"product": "Keyboard", "qty": 1}
                ]
            }
        ]
    }
]

schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("status", StringType(), True),
    StructField("name", StructType([
        StructField("first", StringType(), True),
        StructField("last", StringType(), True)
    ]), True),
    StructField("address", StructType([
        StructField("street", StringType(), True),
        StructField("city", StringType(), True),
        StructField("zip", StringType(), True)
    ]), True),
    StructField("orders", ArrayType(StructType([
        StructField("order_id", StringType(), True),
        StructField("amount", IntegerType(), True),
        StructField("items", ArrayType(StructType([
            StructField("product", StringType(), True),
            StructField("qty", IntegerType(), True)
        ])), True)
    ])), True)
])

df = spark.createDataFrame(json_data, schema=schema)

df_exploded = df.select(
    col("customer_id"),
    col("status"),
    col("name.first").alias("first"),
    col("name.last").alias("last"),
    col("address.street"),
    col("address.city"),
    col("address.zip"),
    explode(col("orders")).alias("order")
)

df_flat = df_exploded.select(
    col("customer_id"),
    col("status"),
    col("first"),
    col("last"),
    col("street"),
    col("city"),
    col("zip"),
    col("order.order_id"),
    col("order.amount"),
    explode(col("order.items")).alias("item")
).select(
    col("customer_id"),
    col("status"),
    col("first"),
    col("last"),
    col("street"),
    col("city"),
    col("zip"),
    col("order_id"),
    col("amount"),
    col("item.product"),
    col("item.qty")
)

display(df_flat)
```
