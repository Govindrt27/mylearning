# Introduction

## What is an Activity?

An **Activity** is the smallest unit of work in an Azure Data Factory (ADF) pipeline.

Example:

```text
Pipeline
│
├── Copy Activity
├── Stored Procedure Activity
├── If Condition Activity
└── Web Activity
```

A pipeline consists of one or more activities that work together to perform ETL/ELT operations.

# 1. Data Movement Activities

## Copy Activity

### Purpose

Copies data from one supported data store to another.

### Common Sources

- Azure SQL Database
- SQL Server
- Oracle
- SAP
- Snowflake
- Azure Blob Storage
- Azure Data Lake Storage (ADLS)
- Amazon S3
- REST API
- Salesforce
- Cosmos DB

### Common Destinations

Any supported Azure Data Factory connector.

### Real-Time Use Case

Every night, copy customer records from SQL Server to Azure Data Lake Storage.

```text
SQL Server
      │
      ▼
Copy Activity
      │
      ▼
Azure Data Lake Storage
```

### Example

**Source**

```sql
dbo.Customers
```

**Destination**

```text
adls/raw/customers/
```

# 3. Control Flow Activities

## If Condition Activity

### Purpose

Execute activities based on a True or False condition.

### Example

```text
File Exists?
      │
 ┌────┴────┐
 │         │
Yes        No
 │          │
 ▼          ▼
Copy     Send Alert
```

### Real-Time Use Case

Only process today's file if it exists.

---

## Switch Activity

### Purpose

Execute different branches depending on a value.

### Example

```text
Country
   │
   ├── USA → Azure SQL
   ├── India → Synapse
   └── UK → Blob Storage
```

### Real-Time Use Case

Different countries require different destination systems.

---

## ForEach Activity

### Purpose

Loop through multiple items.

### Real-Time Use Case

Process hundreds of CSV files automatically.

Example:

```text
Sales_Jan.csv
Sales_Feb.csv
Sales_Mar.csv
...
```

Pipeline:

```text
ForEach
   │
   ▼
Copy Activity
```

---

## Until Activity

### Purpose

Repeat execution until a condition becomes true.

### Real-Time Use Case

Wait until another pipeline completes.

Example:

```text
Status != Completed
        │
        ▼
Retry
        │
        ▼
Completed
```

---

## Wait Activity

### Purpose

Pause the pipeline for a specified duration.

### Example

```text
Wait

5 Minutes
```

### Real-Time Use Case

Wait for a database backup to finish before starting data ingestion.

---

## Execute Pipeline Activity

### Purpose

Call another pipeline.

### Real-Time Use Case

Master pipeline executes multiple child pipelines.

```text
Master Pipeline
      │
      ├── Customer Pipeline
      ├── Sales Pipeline
      └── Finance Pipeline
```

---

## Set Variable Activity

### Purpose

Assign a value to a variable.

### Example

```text
FileName = Sales_2025.csv
```

---

## Append Variable Activity

### Purpose

Append values into an array variable.

### Example

```text
Files

↓

Customer.csv

↓

Sales.csv

↓

Orders.csv
```

---

## Filter Activity

### Purpose

Filter elements from an array.

### Example

Input

```text
100
200
300
400
```

Condition

```text
Value > 200
```

Output

```text
300
400
```

---

## Fail Activity

### Purpose

Explicitly fail the pipeline.

### Real-Time Use Case

Business validation fails.

Example

```text
SalesAmount < 0

↓

Fail Activity
```

# End-to-End Real-Time ETL Example

## Scenario

An e-commerce company receives daily sales files from multiple regions.

### Pipeline Flow

1. **Get Metadata Activity**
   - Check whether today's sales file exists.

2. **If Condition Activity**
   - Continue only if the file exists.

3. **Copy Activity**
   - Copy raw files from Azure Blob Storage to Azure Data Lake Storage.

4. **Mapping Data Flow**
   - Remove duplicates
   - Clean invalid records
   - Calculate profit
   - Standardize formats

5. **Lookup Activity**
   - Retrieve the last processed transaction ID.

6. **Stored Procedure Activity**
   - Update audit tables.

7. **ForEach Activity**
   - Process files from multiple regions.

8. **Delete Activity**
   - Remove temporary staging files.

9. **Web Activity**
   - Notify downstream systems that processing completed.

### Overall Architecture

```text
Blob Storage
      │
      ▼
Get Metadata
      │
      ▼
If Condition
      │
      ▼
Copy Activity
      │
      ▼
Data Flow
      │
      ▼
Lookup
      │
      ▼
Stored Procedure
      │
      ▼
ForEach
      │
      ▼
Delete
      │
      ▼
Web Activity
```

# Complete Real-World ETL Example

## Requirement

Load daily sales data from SAP to Synapse.

### Pipeline Flow

```text
1. Get Metadata
      ↓
2. Validation Activity
      ↓
3. Lookup Control Table
      ↓
4. ForEach Source Table
      ↓
5. Copy Activity
      ↓
6. Mapping Data Flow
      ↓
7. Stored Procedure
      ↓
8. Azure Function (Send Notification)
      ↓
9. Success/Fail Email
```

### Business Result

* Fully automated ingestion
* Incremental loading
* Data quality validation
* Reporting-ready data warehouse

This architecture is commonly used in banking, retail, insurance, healthcare, manufacturing, and telecom projects running on Azure Data Factory.
