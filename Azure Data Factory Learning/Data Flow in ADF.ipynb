**Data Flow in Azure Data Factory (ADF)** is a **visual, no-code/low-code data transformation service** that allows you to transform, clean, aggregate, join, and enrich data without writing Spark code. Under the hood, ADF executes Data Flows on managed **Apache Spark clusters**. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview), [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

### Why use Data Flow?

Data Flows are used when you need to perform ETL/ELT transformations such as:

* Filtering records
* Joining multiple datasets
* Aggregations (SUM, COUNT, AVG, etc.)
* Data cleansing and validation
* Derived columns
* Data type conversions
* Sorting and windowing functions
* Upserts into target systems [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview), [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview)

### Basic Architecture

```text
Source
   ↓
Transformations
   ↓
Sink
```

Example:

```text
SQL Database
      ↓
   Filter
      ↓
 Derived Column
      ↓
    Join
      ↓
Azure Data Lake
```

### Main Components

1. **Source**
   * Reads data from SQL Server, Azure SQL DB, ADLS, Blob Storage, Synapse, etc. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview)

2. **Transformations**
   Common transformations include:
   * Filter
   * Join
   * Aggregate
   * Derived Column
   * Lookup
   * Union
   * Conditional Split
   * Window
   * Pivot/Unpivot [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

3. **Sink**
   * Writes transformed data to the destination system such as SQL Database, ADLS, Blob Storage, Synapse Analytics, etc. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview), [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview)

### Types of Data Flows in ADF

#### 1. Mapping Data Flow

* Visual drag-and-drop design.
* Most commonly used.
* Generates Spark code automatically. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview)

#### 2. Wrangling Data Flow

* Power Query-based transformations.
* Intended for self-service data preparation. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview)

### Difference Between Copy Activity and Data Flow

| Feature                | Copy Activity | Data Flow         |
| ---------------------- | ------------- | ----------------- |
| Data movement          | Yes           | Yes               |
| Data transformation    | Limited       | Extensive         |
| Spark execution        | No            | Yes               |
| Visual transformations | No            | Yes               |
| Best for               | Moving data   | Transforming data |

### Real-World Example

Suppose you have:

* Customer data in SQL Server
* Product data in Azure Data Lake

Requirements:

1. Filter active customers.
2. Join customer and product data.
3. Calculate total sales.
4. Load results into Synapse.

You would create:

```text
Source(Customer)
         \
          Join → Aggregate → Sink(Synapse)
         /
Source(Product)
```

# Top 20 ADF Data Flow Interview Questions and Answers

## 1. What is a Data Flow in Azure Data Factory?

**Answer:**
A Data Flow is a visual data transformation service in Azure Data Factory that enables users to build ETL/ELT transformations without writing Spark code. Data Flows execute on ADF-managed Apache Spark clusters and are typically used for data cleansing, aggregation, joins, filtering, and transformations. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview), [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

## 3. What is the difference between Copy Activity and Data Flow?

**Answer:**

| Copy Activity                  | Data Flow                         |

| Moves data                     | Transforms data                   |

## 4. What are the main components of a Mapping Data Flow?

**Answer:**

A Mapping Data Flow consists of:

1. Source
2. Transformations
3. Sink

```text
Source → Transformations → Sink
```

Common transformations include Join, Aggregate, Filter, Derived Column, Lookup, and Conditional Split. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview), [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview)

***

## 5. What is a Source Transformation?

**Answer:**

Source Transformation is the starting point of a Data Flow where data is read from a source system such as:

* Azure SQL Database
* SQL Server
* Azure Data Lake Storage
* Blob Storage
* Synapse Analytics

It defines schema and data ingestion settings. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/concepts-data-flow-overview)

***

## 6. What is a Sink Transformation?

**Answer:**

Sink Transformation writes processed data to a target system.

Examples:

* Azure SQL Database
* ADLS Gen2
* Blob Storage
* Synapse Analytics

A Data Flow can have multiple sinks. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

***

## 7. What is Schema Drift?

**Answer:**

Schema Drift allows Data Flow to process columns that are not explicitly defined in the source schema.

Benefits:

* Handles changing source schemas.
* Avoids pipeline failures when new columns appear.
* Useful for semi-structured data such as JSON.

***

## 8. What is a Derived Column Transformation?

**Answer:**

Derived Column is used to:

* Create new columns
* Modify existing columns
* Apply business rules

Example:

```sql

```

 [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

## 9. What is Filter Transformation?

**Answer:**

Filter Transformation removes unwanted rows based on specified conditions.

Example:

```sql
Salary > 50000
```

Only qualifying records move to the next transformation. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

***

## 10. What is Conditional Split?

**Answer:**

Conditional Split routes records to different streams based on conditions.

Example:

```text
Salary > 100000 → High Salary Stream
Salary <=100000 → Normal Salary Stream
```

Useful for branching business logic. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

***

## 11. What is the difference between Join and Lookup?

**Answer:**

### Join

* Combines two large datasets.
* Supports Inner, Left, Right, Full joins.

### Lookup

* Used to fetch reference data.
* Usually used when one dataset is smaller. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

***

## 14. What is Alter Row Transformation?

**Answer:**

Alter Row specifies database operations on incoming records.

Used for:

- Insert

- Delete

## 15. What is the difference between Pipeline Parameters and Data Flow Parameters?

**Answer:**

### Pipeline Parameters
- Defined at pipeline level
- Passed into activities

### Data Flow Parameters
- Defined within Data Flow
- Used inside transformation logic

Example:

```text
Pipeline Parameter
        ↓
Data Flow Parameter
        ↓
Filter Logic
````

***

## 16. What is Data Preview in Data Flow?

**Answer:**

Data Preview allows developers to:

* View sample records
* Validate transformations
* Debug expressions
* Test logic before execution

It requires a debug cluster to be running.

***

## 17. What is Debug Mode in Data Flow?

**Answer:**

Debug Mode provisions a temporary Spark cluster that allows:

* Data preview
* Expression validation
* Transformation testing
* Real-time troubleshooting

Without debug mode, preview data is unavailable.

***

## 18. What is a Surrogate Key Transformation?

**Answer:**

Surrogate Key generates unique sequential keys for records.

Example:

```text
CustomerID
-----------
1
2
3
4
```

Commonly used in data warehouse dimensions.

***

## 19. How does ADF Data Flow handle Slowly Changing Dimension (SCD)?

**Answer:**

SCD implementation typically uses:

* Lookup transformation
* Alter Row transformation
* Conditional Split
* Upsert logic

### SCD Type 1

Overwrites old data.

### SCD Type 2

Maintains history by adding new records and expiration dates.

***

## 20. How can you optimize performance in ADF Data Flow?

**Answer:**

Best practices:

1. Use partitioning effectively.
2. Filter data as early as possible.
3. Avoid unnecessary columns.
4. Use broadcast joins for smaller lookup tables.
5. Enable schema projection.
6. Cache reusable datasets.
7. Optimize sink writes.
8. Use proper cluster sizing.

***

# Important Scenario-Based Interview Question

### Q: How would you load incremental data using ADF Data Flow?

**Answer:**

1. Store the last successful load timestamp.
2. Pass timestamp as a parameter.
3. Filter source records:

```sql
ModifiedDate > LastLoadDate
```

4. Use Alter Row for inserts/updates.
5. Write to target.

This minimizes data movement and improves performance.

***

# Frequently Asked Advanced Questions

### What is Broadcast Join?

A performance optimization where a small dataset is copied to all Spark worker nodes to avoid expensive shuffles.

### What is Sink Ordering?

Determines execution order when multiple sinks exist in a Data Flow.

### Can Data Flow process semi-structured data?

Yes. JSON, Avro, Parquet, XML, and hierarchical data can be processed using transformations such as Flatten and Parse. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

### What is Flatten Transformation?

Used to expand nested arrays and hierarchical structures into relational rows. [\[learn.microsoft.com\]](https://learn.microsoft.com/en-us/azure/data-factory/data-flow-transformation-overview)

***

## Interview Tip

For experienced ADF developers (3–8 years), interviewers frequently focus on:

* Schema Drift
* SCD Type 2
* Broadcast vs Shuffle Join
* Alter Row and Upsert
* Incremental Load Design
* Data Flow Performance Tuning
* Parameterization
* Debugging Failures
* Spark Partitioning
* End-to-End ETL Design

Master these topics and you'll be well prepared for most ADF Data Flow interviews.
