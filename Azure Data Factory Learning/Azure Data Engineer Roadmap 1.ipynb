# Capstone Project

## Enterprise Sales Data Platform

### Data Sources

- Azure SQL Database
- REST API
- CSV Files (ADLS)
- ERP System

### Workflow

1. Ingest data using Azure Data Factory.
2. Store raw data in ADLS Bronze.
3. Transform data with Databricks (PySpark).
4. Save curated data as Delta Lake Silver.
5. Build business-ready Gold tables.
6. Govern access with Unity Catalog.
7. Deploy pipelines using Azure DevOps CI/CD.
8. Monitor jobs with Azure Monitor and Databricks logs.
9. Visualize Gold data in Power BI.

### Deliverables

- Azure Data Factory pipelines
- Databricks notebooks
- Delta Lake tables
- Unity Catalog governance
- Azure DevOps YAML pipelines
- Documentation
- Monitoring dashboards
