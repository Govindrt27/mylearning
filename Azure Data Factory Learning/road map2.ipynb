# 1. Delta Lake (Must Learn)

## What is Delta Lake?

Delta Lake is an open storage layer built on top of Data Lakes (ADLS) that provides:

* ACID Transactions
* Schema Enforcement
* Schema Evolution
* Time Travel
* Data Versioning
* Upserts and Deletes

Think of it as:

```text
Parquet + Reliability + Transactions
```

***

## Delta Architecture

```text
ADLS
  |
  +-- Delta Table
          |
          +-- Parquet Files
          +-- _delta_log
```

The `_delta_log` folder stores transaction history.

***

## Delta Lake Topics to Learn

### Create Delta Table

```sql
CREATE TABLE customer
USING DELTA
AS SELECT * FROM source
```

***

### Insert

```sql
INSERT INTO customer
VALUES (1,'Govind')
```

***

### Update

```sql
UPDATE customer
SET name='Govind Kumar'
WHERE id=1
```

***

### Delete

```sql
DELETE FROM customer
WHERE id=1
```

***

### Merge (Most Important)

Interviewers love this.

```sql
MERGE INTO target t
USING source s
ON t.id=s.id

WHEN MATCHED
THEN UPDATE SET *

WHEN NOT MATCHED
THEN INSERT *
```

### Use Cases

* Incremental Loads
* SCD Type 1
* SCD Type 2
* CDC processing

***

## Time Travel

Allows querying historical versions.

```sql
SELECT *
FROM customer VERSION AS OF 5
```

Useful for:

* Audit
* Data recovery
* Root cause analysis

***

## Optimize

Improves query performance.

```sql
OPTIMIZE customer
```

***

## Z-Ordering

```sql
OPTIMIZE customer
ZORDER BY (customer_id)
```

Use when filtering frequently on specific columns.

***

## Vacuum

Removes old files.

```sql
VACUUM customer
```

Interview Question:

**Why run VACUUM after OPTIMIZE?**

Answer:

* OPTIMIZE creates larger compacted files.
* VACUUM removes obsolete small files.

***

# Delta Lake Interview Questions

### What are Delta Lake advantages over Parquet?

Answer:

| Parquet         | Delta                 |
| --------------- | --------------------- |
| No ACID         | ACID Transactions     |
| No Time Travel  | Time Travel           |
| Limited Updates | Update/Delete Support |
| No Merge        | Merge Support         |
| No Versioning   | Versioning            |

***

### What is Delta Log?

Answer:

Transaction history maintained inside:

```text
_delta_log
```

Contains:

* commits
* metadata
* schema changes
* versions

***

### What is Time Travel?

Answer:

Ability to query older snapshots of data.

```sql
VERSION AS OF
```

# Managed Tables vs External Tables

## Managed Table

```text
Databricks owns data
```

Dropping table removes data.

```sql
DROP TABLE customer
```

Deletes underlying files.

***

## External Table

```text
ADLS owns data
```

Dropping table removes metadata only.

Data remains in Storage Account.

Interview favorite question.

# Unity Catalog Interview Questions

## What is Unity Catalog?

Answer:

Unity Catalog is Databricks' centralized governance solution that provides unified access control, auditing, lineage, and data discovery across all Databricks workspaces.

***

## What is a Metastore?

Answer:

A Metastore is the top-level container in Unity Catalog that stores metadata about catalogs, schemas, tables, permissions, and data assets.

***

## Explain Unity Catalog hierarchy.

Answer:

```text
Metastore
 ↓
Catalog
 ↓
Schema
 ↓
Tables
```

***

## Difference Between Hive Metastore and Unity Catalog?

### Hive Metastore

```text
Workspace Level
Limited Governance
```

### Unity Catalog

```text
Account Level
Centralized Governance
Cross Workspace
Fine-Grained Security
```

***

## What are External Locations?

Answer:

External Locations provide secure access from Unity Catalog to cloud storage locations such as ADLS.

***

## What is Data Lineage?

Answer:

Lineage tracks data movement from source to destination and helps identify downstream impacts of changes.
