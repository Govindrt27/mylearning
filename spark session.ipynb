# How to Create SparkSession + Common PySpark Imports (Real-Time Guide)

In PySpark, **SparkSession** is the entry point to work with Spark DataFrames, SQL, and structured data APIs. Without it, you cannot read data or perform transformations.

---

# 1. How to Create SparkSession

## Basic SparkSession

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MySparkApp") \
    .getOrCreate()
```

---

## Databricks / Cluster Mode (Best Practice)

In Databricks or managed clusters, SparkSession is already created:

```python
spark
```

No need to explicitly create it.

---

## SparkSession with Configurations

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ETL_Pipeline") \
    .config("spark.sql.shuffle.partitions", "200") \
    .config("spark.executor.memory", "4g") \
    .config("spark.driver.memory", "2g") \
    .getOrCreate()
```

---

## Enable Hive Support (Very Important in Real Projects)

```python
spark = SparkSession.builder \
    .appName("HiveIntegration") \
    .enableHiveSupport() \
    .getOrCreate()
```

Used when working with Hive tables.

---

## Stop SparkSession

```python
spark.stop()
```

# 4. Real-Time Example (ETL Job Setup)

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, coalesce
from pyspark.sql.window import Window
from pyspark.sql.types import *

spark = SparkSession.builder \
    .appName("Customer_ETL") \
    .getOrCreate()

df = spark.read.csv("/data/customers.csv", header=True)

df = df.withColumn("city", coalesce(col("city"), lit("Unknown")))
```

# 5. Key Interview Points

### Q1. What is SparkSession?
SparkSession is the **single entry point** to Spark functionality for DataFrame and SQL APIs.

---

### Q2. Why do we use `getOrCreate()`?
It avoids creating multiple Spark sessions and reuses an existing one if available.

---

### Q3. Why is `pyspark.sql.functions` important?
Because most transformations (filtering, aggregations, null handling, string manipulation) are done using these functions.

---

### Q4. Why is Window function important?
Used for:
- Ranking
- Deduplication
- Latest record selection
- Time-based analysis

---

# 6. Best Practices

- Always create SparkSession as the first step in scripts
- Use `pyspark.sql.functions` instead of Python loops
- Import only required functions in production (avoid `*` in large projects)
- Use `Window` for complex deduplication logic
- Enable Hive support if working with Hive tables
